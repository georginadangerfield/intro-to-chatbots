# The Ethics of Chatbots

Ethics should be at the forefront of how Artifical Intelligence (AI) is used. Regardless of whether this AI is driverless cars, social media moderation, healthcare or virtual assistants, failing to take ethics into consideration can lead to unhappiness and distrust among its users.

Chatbots that are ethical and implemented in a responsible way will do an amazing job of promoting brand loyalty and encouraging relationships built on trust. So what ethical concerns do we need to think about in order to make this happen?

## 1. Privacy


It's extremely important to consider user privacy when creating chatbots. The developer will need to ensure the security of the data that the bot will store, to avoid exposing potentially sensitive information to the general public. Users won't ineract with a chatbot if there is even a shadow of a doubt that the questions they ask will not be private and secure.

## 2. Identity

Another thing to think about is the identity of the chatbot, i.e. its gender, race and nationality. Should the bot have a gender, or should it be genderless? The answer to this question might differ depending on the language it will be conversing in, as a genderless bot might struggle with Spanish but would be fine with English. 

On the other hand, many existing chatbots can be described as female, but this has the potential to encourage stereotypes of women being submissive and compliant. According to  UCLA professor Safiya Noble, chatbots can "function as powerful socialization tools, and teach people, in particular children, about the role of women, girls, and people who are gendered female to respond on demand."

And what about race and nationality? It may be quite difficult to create a chatbot that is doesn't have any form of racial or ethnic identity without reinforcing dominant narratives about race and ethnicity that are probably already present in the user's country.

## 3. Trust and Transparency

When talking about transparency in this context, we simply mean that the user should be able to tell whether they are speaking to a human or a bot. People can usually tell the difference, but as AI becomes more and more convincing and human-like, it's essential that developers are honest about it and don't pretend that the machines are humans, in order to build trust between user and company. The option to speak to a real person should also be in place if the chatbot isn't able to give them the service or response that they are after.

## 4. Proper Training

We also need to take caution when training bots to ensure that it will behave appropriately when interacting with users. If not trained properly, the chatbot can potentially start to use racist, sexist or abusive language. This is what happened to [Tay](https://en.wikipedia.org/wiki/Tay_%28bot%29), the bot that Microsoft created for use on Twitter that would respond based on how Twitter users interacted with it. Conversations started off friendly, but it only took 16 hours for Tay to be taken down as trolls started to use language full of rasism, sexism and other offensive content which Tay immediately imitated.
